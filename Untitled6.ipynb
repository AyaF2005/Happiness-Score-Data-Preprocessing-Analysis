```python
# Sort by 'Happiness Score' in descending order
df = df.sort_values(by='Happiness Score', ascending=False)

# Assign rank based on sorted happiness score
df['Happiness Rank'] = df['Happiness Score'].rank(ascending=False, method='first').astype(int)
```

```python
df['health_zscore'] = (df['Health (Life Expectancy)'] - df['Health (Life Expectancy)'].mean()) / df['Health (Life Expectancy)'].std()

# Identify outliers
df['outlier'] = df['health_zscore'].apply(lambda x: abs(x) >= 3)

# Display outliers
outliers = df[df['outlier'] == True].sort_values(by="health_zscore", ascending=False)

print("Outliers:")
print(outliers)
```

```python

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['Health (Life Expectancy)'].quantile(0.25)
Q3 = df['Health (Life Expectancy)'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
df['outlier_iqr'] = df['Health (Life Expectancy)'].apply(lambda x: x < lower_bound or x > upper_bound)

# Display outliers
outliers_iqr = df[df['outlier_iqr'] == True].sort_values(by="Health (Life Expectancy)", ascending=False)

print("Outliers using IQR method:")
print(outliers_iqr)
```
